```python
def get_peak_heigthts(data, peaks):
    peak_heights = []
    for i in range(0, len(peaks['left_edges'])):
        peak_heights.append(data[peaks['left_edges'][i]])
    return np.array(peak_heights)
```

```python
def intergrals(data, peaks):
    integrals = []
    for i in np.arange(0, len(peaks['left_edges']), 1):
        l = int(np.ceil(peaks['left_ips'][i]))
        r = int(np.ceil(peaks['right_ips'][i]))
        I = np.sum(data[l:r])
        integrals.append(I)
    return np.array(integrals)

def promptFraction(data, peaks, prompt_right_boundary=30):
    integrals, PF = [], []
    for i in np.arange(0, len(peaks['left_edges']), 1):
        l = int(np.ceil(peaks['left_ips'][i]))
        r = int(np.ceil(peaks['right_ips'][i]))
        rprompt = int(np.ceil(peaks['left_ips'][i])+prompt_right_boundary)
        P = np.sum(data[l:rprompt])
        I = np.sum(data[l:r])
        PF.append(P/I)
        integrals.append(I)
    return np.array(PF), np.array(integrals)
```

```python
prompt_boundary=50
n_PF, n_I = promptFraction(n_data_flat, n_peaks, prompt_right_boundary=prompt_boundary)
gn_PF, gn_I = promptFraction(gn_data_flat, gn_peaks, prompt_right_boundary=prompt_boundary)
c_PF, c_I = promptFraction(c_data_flat, c_peaks, prompt_right_boundary=prompt_boundary)
```
