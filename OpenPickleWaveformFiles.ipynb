```python
import os
import sys
import numpy as np
import matplotlib.pyplot as plt
# from scipy.signal import find_peaks
import pickle
```

```python
with open(f'PNS_waveforms.pkl', 'rb') as file:
    waveforms = pickle.load(file)
```

```python
fig, ax = plt.subplots(1, 1, figsize=(16, 6))
for waveform in waveforms:
    ax.plot(waveform.transpose())
```

```python
fig, ax = plt.subplots(1, 1, figsize=(16, 6))
for waveform in waveforms[:1000]:
    ax.plot(waveform.transpose())
```

```python
integrals = []
for waveform in waveforms[:1000]:
    integrals.append(waveform.sum())
```

```python
plt.hist(integrals, bins=100)
```
